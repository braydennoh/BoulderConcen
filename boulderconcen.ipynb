```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
import os

# Function to process sediment data from a file
def process_sediment_data(filename):
    sediment_data = []
    current_sediment = {}

    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('c:'):
                if 'c' not in current_sediment:
                    current_sediment['c'] = []
                current_sediment['c'].append(float(line.split(':')[1].strip()))
            elif line.startswith('r:'):
                if 'r' not in current_sediment:
                    current_sediment['r'] = []
                current_sediment['r'].append(float(line.split(':')[1].strip()))
            elif line == '':
                if current_sediment:
                    sediment_data.append(current_sediment)
                    current_sediment = {}

    # Define the lists to store the values
    ax1_values, ax2_values, ax3_values = [], [], []

    # Extract the data from the sediment_data list
    for sediment in sediment_data:
        ax1_values.append(sediment['r'][0])
        ax2_values.append(sediment['r'][1])
        ax3_values.append(sediment['r'][2])

    # Function to calculate the major axis length
    def calculate_major_axis_length(ax1, ax2, ax3):
        return [math.sqrt(ax1[i]**2 + ax2[i]**2 + ax3[i]**2) for i in range(len(ax1))]

    # Calculate the diameters
    diameter = calculate_major_axis_length(ax1_values, ax2_values, ax3_values)

    # Filter values
    filtered_indices = [i for i, d in enumerate(diameter) if 0.25 <= d <= 5]
    filtered_diameters = [diameter[i] for i in filtered_indices]

    return filtered_diameters

# Initialize lists to store D30, D50, and D90 values for each file
D30_values = []
D50_values = []
D90_values = []

plt.figure(figsize=(10, 6))

# Loop through each file and process the data
for i in range(1, 14):  # For mont1 to mont13
    filename = f'mont{i}.txt'
    filtered_diameters = process_sediment_data(filename)
    sorted_diameters = np.sort(filtered_diameters)
    cdf = np.arange(1, len(sorted_diameters) + 1) / len(sorted_diameters)

    # Calculate D30, D50, D90 for the current distribution
    D30 = np.percentile(sorted_diameters, 30)
    D50 = np.percentile(sorted_diameters, 50)
    D90 = np.percentile(sorted_diameters, 90)

    # Store the values in their respective lists
    D30_values.append(D30)
    D50_values.append(D50)
    D90_values.append(D90)

    # Plot the cumulative density function for each distribution
    plt.plot(sorted_diameters, cdf, label=f'mont{i}')

# Calculate the average values for D30, D50, and D90
average_D30 = np.mean(D30_values)
average_D50 = np.mean(D50_values)
average_D90 = np.mean(D90_values)

# Print out the average values
print(f'Average D30: {average_D30:.2f}')
print(f'Average D50: {average_D50:.2f}')
print(f'Average D90: {average_D90:.2f}')

# After computing the average values of D30, D50, and D90, we will plot them on the graph

# Plot the average D30, D50, and D90 lines
plt.axhline(y=0.3, color='blue', linestyle='--', label=f'Average D30 = {average_D30:.2f}')
plt.axhline(y=0.5, color='green', linestyle='--', label=f'Average D50 = {average_D50:.2f}')
plt.axhline(y=0.9, color='purple', linestyle='--', label=f'Average D90 = {average_D90:.2f}')

# Print the average D30, D50, and D90 values on the plot
plt.text(sorted_diameters[-1], 0.3, f'{average_D30:.2f}', verticalalignment='bottom', color='blue')
plt.text(sorted_diameters[-1], 0.5, f'{average_D50:.2f}', verticalalignment='bottom', color='green')
plt.text(sorted_diameters[-1], 0.9, f'{average_D90:.2f}', verticalalignment='bottom', color='purple')
plt.xlim([0,2.5])
plt.ylim([0,1.1])

# Add labels and legend
plt.xlabel('Sediment diameter (m)')
plt.ylabel('Cumulative density')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
import os
from scipy.spatial import Delaunay  # Import Delaunay class from scipy.spatial

output_dir = '/Users/braydennoh/Downloads/debrisflow'

sediment_data = []
current_sediment = {}

with open('aguupstream1.txt', 'r') as file:    
    for line in file:
        line = line.strip()
        if line.startswith('c:'):
            if 'c' not in current_sediment:
                current_sediment['c'] = []
            current_sediment['c'].append(float(line.split(':')[1].strip()))
        elif line.startswith('r:'):
            if 'r' not in current_sediment:
                current_sediment['r'] = []
            current_sediment['r'].append(float(line.split(':')[1].strip()))
        elif line == '':
            if current_sediment:
                sediment_data.append(current_sediment)
                current_sediment = {}

x_values = []
y_values = []
z_values = []
ax1_values = []
ax2_values = []
ax3_values = []

for sediment in sediment_data:
    x_values.append(sediment['c'][0])
    y_values.append(sediment['c'][1])
    z_values.append(sediment['c'][2])
    ax1_values.append(sediment['r'][0])
    ax2_values.append(sediment['r'][1])
    ax3_values.append(sediment['r'][2])

def calculate_major_axis_length(x, y, z):
    major_axis_lengths = []
    for i in range(len(x)):
        major_axis_length = math.sqrt(x[i]**2 + y[i]**2 + z[i]**2)
        major_axis_lengths.append(major_axis_length)
    return major_axis_lengths

x_dimensions = ax1_values
y_dimensions = ax2_values
z_dimensions = ax3_values
diameter = calculate_major_axis_length(ax1_values, ax2_values, ax3_values)
 
# Filter values less than or equal to 5 and greater than or equal to 0.25
filtered_x_values = [x for i, x in enumerate(x_values) if diameter[i] <= 5 and diameter[i] >= 0.25]
filtered_y_values = [y for i, y in enumerate(y_values) if diameter[i] <= 5 and diameter[i] >= 0.25]
filtered_z_values = [z for i, z in enumerate(z_values) if diameter[i] <= 5 and diameter[i] >= 0.25]
filtered_diameters = [d for d in diameter if d <= 5 and d >= 0.25]
medianvaluediameter = np.median(filtered_diameters)
p_s = 2
p = 1
g = 9.8
S = 0.1
t_star = 0.06
w = 10  # width for area calculation

# Initialize lists to store calculated values for each iteration
median_lengths = []
A_t_list = []
A_IP_list = []
A_m_list = []
Am_At_ratio_list = []
q_star_final_list = []
q_star_list = []
t = t_star * (p_s - p) * g * np.array(filtered_diameters)  # Convert to NumPy array
r_values = t / (p * g * S)

for i in np.arange(0, 3, 0.1):
    iterated_indices = np.where(r_values > i)[0]
    iterated_x_values = np.array(filtered_x_values)[iterated_indices]  # Convert to NumPy array
    iterated_y_values = np.array(filtered_y_values)[iterated_indices]  # Convert to NumPy array
    iterated_r_values = np.array(r_values)[iterated_indices]  # Convert to NumPy array

    # Convert iterated_r_values back to diameters
    iterated_d_values = iterated_r_values / ((t_star * (p_s - p) * g) / (p * g * S))
    median_diameter = np.median(iterated_d_values)
    tau_iterated = (p * g * S * i)
    tau_star_iterated = tau_iterated / ((p_s - p) * g * median_diameter)
    q_star = 5.7*(tau_iterated - tau_star_iterated)**1.5
    # Check if there are enough points for Delaunay triangulation
    iterated_points = np.vstack((iterated_x_values, iterated_y_values)).T
    if iterated_points.shape[0] < 4:
        median_lengths.append(np.nan)
        continue  # Skip the rest of the loop if not enough points

    # Perform the Delaunay triangulation
    iterated_tri = Delaunay(iterated_points)
    
    # Calculate the lengths of the edges of the triangles
    iterated_edges = iterated_points[iterated_tri.simplices]
    iterated_line_lengths = np.sqrt(np.sum(np.diff(iterated_edges, axis=1)**2, axis=2).flatten())
    
    # Calculate the median length of lines not greater than 10
    iterated_median_length = np.median(iterated_line_lengths[(iterated_line_lengths <= 100) & (iterated_line_lengths > 0.1)])
    median_lengths.append(iterated_median_length)
    
    # Calculate areas based on the median length and median diameter
    A_t = (np.sqrt(3) * iterated_median_length / 2) * w
    A_IP = (np.pi * median_diameter**2 * w) / (4 * iterated_median_length)
    A_m = A_t - A_IP
    q_star_final = q_star*(A_m /A_t)
    # Store the calculated areas
    A_t_list.append(A_t)
    A_IP_list.append(A_IP)
    A_m_list.append(A_m)
    Am_At_ratio_list.append(A_m /A_t)
    q_star_final_list.append(q_star_final)
    q_star_list.append(q_star)
    marker_sizes = np.sqrt(iterated_r_values) * 50
    plt.figure(figsize=(3, 2.5))
    plt.scatter(iterated_x_values, iterated_y_values, c=iterated_r_values, cmap='viridis', s=marker_sizes, vmin=0, vmax=3)
    plt.triplot(iterated_points[:, 0], iterated_points[:, 1], iterated_tri.simplices.copy(), color="black",lw=0.7)
    plt.colorbar(label='Flow Depth (m)')
    plt.xlabel('(m)')
    plt.ylabel('(m)')
    plt.title(f'Iteration: {i:.1f}')
    plt.gca().set_axis_off()
    plt.savefig(os.path.join(output_dir, f'delaunay_iteration_{i}.png'))
    plt.close()
```
